```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df1=pd.read_csv('O_mom_sort', delim_whitespace=True, header=None)
df2=pd.read_csv('/Users/arup/Documents/Urmi_IACS/results/DRS/vac-Li0.2/UD3/fix_U/O_mom_sort', delim_whitespace=True, header=None)
df3=pd.read_csv('/Users/arup/Documents/Urmi_IACS/results/DRS/vac-Li0.2/Fixed-TM/UD3/O_mom_sort', delim_whitespace=True, header=None)
#print(df1[2])
#df11=df1.iloc[0:18]
O_mom_1 = np.array(df1[2], dtype=np.float32)
O_mom_2 = np.array(df2[2], dtype=np.float32)#,label='Li0.2-Re')
O_mom_3 = np.array(df3[2], dtype=np.float32)#,label='Li0.2-Fixed')

fig, axes=plt.subplots()

parts=axes.violinplot(dataset=[O_mom_1,O_mom_2,O_mom_3],showmeans=True,showextrema=True)
# Custom colors for each violin
colors = ['skyblue', 'lightgreen', 'salmon']
for i, pc in enumerate(parts['bodies']):
    pc.set_facecolor(colors[i])
    pc.set_edgecolor('black')
    pc.set_alpha(0.8)

# Optional: style mean/median lines
for partname in ['cmeans', 'cbars', 'cmins', 'cmaxes']:
    if partname in parts:
        parts[partname].set_edgecolor('black')
        parts[partname].set_linewidth(1)

axes.set_xticks([1, 2, 3])
# axes.set_xticklabels(['O_mom_1', 'O_mom_2', 'O_mom_3'])
axes.set_xticklabels(['DRS-Li1.2', 'DRS-Li0.2 (Re)', 'DRS-Li0.2 (Fix)'],fontsize=16)

axes.set_ylabel('Magnetic moment of oxygen ($\mu_B$)',fontsize=16)
#axes.tick_params(labelbottom=False, bottom=False)
#plt.xticks([1, 2, 3], ['O_mom_1', 'O_mom_2', 'O_mom_3'])
axes.set_ylim(0,1.2)
axes.tick_params(axis='both', labelsize=14)
plt.tight_layout()
plt.rcParams['font.family'] = 'Arial'
#plt.grid(True)
#plt.savefig('DRS-O-mom-violin.tif',dpi=500)
plt.savefig('DRS-O-mom-violin.eps',dpi=500)
plt.show()
```
